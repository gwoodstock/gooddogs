```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shelters = pd.read_csv('./datasets/working_data/austin_clean_breed_stats.csv')
```

# Info

```python
shelters.head(2)
```

```python
shelters.info()
```

```python
shelters.describe()
```

Convert to datetimes

```python
shelters['date_in'] = pd.to_datetime(shelters['date_in'])
shelters['date_out'] = pd.to_datetime(shelters['date_out'])

shelters['time_in_shelter'] = pd.to_timedelta(shelters['time_in_shelter'])
shelters['time_in_shelter'] = shelters['time_in_shelter'].dt.days
```

# Exploring the Data

```python
plt.figure(figsize=(16,14))
sns.heatmap(
    shelters.corr()[['time_in_shelter']].sort_values(by='time_in_shelter', ascending=False),
    annot=True,
)

plt.tight_layout();
```

None of the numerical features show a high enough correlation to look into further.

```python
shelters['location'].value_counts().head(20)
```

Most of the locations just give the city with no other details.

# Breeds

### Breed 1

```python
breed_1 = shelters['breed_1'].value_counts().head(20)                               # top 20 breeds for breed_1
breed_1 = pd.DataFrame(breed_1).reset_index()                                       # create data frame and reset index
breed_1.rename(columns={'index': 'breed_1', 'breed_1': 'count'}, inplace=True)      # rename columns
breed_1.head()
```

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')

sns.barplot(
    data=breed_1, 
    orient='h', 
    x='count', 
    y='breed_1',
    ci=None,
    color='green',
    alpha=.85
);
```

Top 5 primary breeds are labs, pit bulls, chihuahua, german shepherds, and australian cattle dogs

### Breed 2

```python
breed_2 = shelters['breed_2'].value_counts().head(20)                               # top 20 breeds for breed_2
breed_2 = pd.DataFrame(breed_2).reset_index()                                       # create data frame and reset index
breed_2.rename(columns={'index': 'breed_2', 'breed_2': 'count'}, inplace=True)      # rename columns
breed_2.head()
```

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')

sns.barplot(
    data=breed_2, 
    orient='h', 
    x='count', 
    y='breed_2',
    ci=None,
    color='green',
    alpha=.85
);
```

Top 5 secondary breeds are pit bulls, chihuahua, labs, german shepherds, and australian cattle dogs

```python
# breed_1 groupby
# looking at time_in_shelter and amount of dogs per breed
breed_gby = shelters.groupby(['breed_1']).agg(['mean','size'])
breed_gby = breed_gby.sort_values(by=('time_in_shelter','mean'), ascending=False)
breed_gby.head(20)
```

Pit bulls was one of the top breeds for time in shelter. They are also one of the most common breeds in the dataset

# Intake Type

```python
shelters['intake_type'].value_counts()
```

```python
intake_gby = shelters.groupby(by='intake_type').mean()
intake_gby.sort_values(by='time_in_shelter')
```

Dogs that were surrendered for the intake type were in the shelter the longest. Maybe they had some issues and that's why they were given up

# Condition

```python
shelters['condition'].value_counts()
```

```python
condition_gby = shelters.groupby(by='condition').mean()
condition_gby.sort_values(by='time_in_shelter')
```

Dogs with behavior issues are in the shelter the longest. There are very few dogs in the dataset with this condition though
